```python
# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable'prodvalue' through boxplot, and write your findings.
```

```python
sns.boxplot(x="year",y="prodvalue", data= df,color = "blue")
plt.title("Relationship between year and Production Value")
```

```python
#FINDINGS: The median being towards the bottom of the box suggests that most product values in those years fall below the median.We can see presence of outliers significantly for the years 2010,2011,2012. 2010 being highest.
```

```python
#There is a strong positive correlation between "numcol" and "totalprod" (0.989). which shows taht with increase in the number of columns,in the total production also tends to increase.
#Same way, "totalprod" has strong positive correlations with "stocks" (0.977) and "prodvalue" (0.971). This indicates that higher total production is associated with higher stock levels and production value.
#There is a very strong negative correlation between "priceperlb" and all other variables except "prodvalue".
#The correlation between "yieldpercol" and other variables is generally weaker and between "prodvalue" and "priceperlb" is also weak.
```
